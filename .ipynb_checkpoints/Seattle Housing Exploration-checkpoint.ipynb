## SOLVING FOR WEIGHTS WITH LINEAR ALGEBRA

its very convenient to be able to get the weights with the following approach.

As mentioned above, since, there's only one feature to consider (sqft), we can solve for the weights with an analytic approach.

Essentially, this means using basic linear algebra techniques, to solve for our weight, weights being our slope and intercept in single-featured example.

The analytic approach is a direct solution.
It does not require 
<ul>gradient descent</ul>
<ul>thousands of iterations</ul>
<ul>learning rate guesswork</ul>

In fact, the entire solution can be solved with an analytic approach. If the code is vectorized, an analytic approach would work up to even 10,000 features. 

```python
def normal_equation(X, y):
    return np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
normal_equation(X_with_ones_col, y)
```

# Lets plot a cost function

### debugging plot function
I'm going to explicitly perform operations before throwing everything into a function, because if I put everything in a function first, I often have a hard time debugging

```python
steps = np.linspace(-3, 3, 5)  # this mesh contains 1.84 and 2.29
w0, w1 = np.meshgrid(steps, steps)
w0_ = w0.ravel()
w1_ = w1.ravel()
w = np.column_stack([w0_, w1_])
w
```

```python
def calculateCostArray(X, y, i):
    return computeCost(X, y, w[i])
```

```python
def computeCost(X, y, w):
    t = X @ w - y
    t_squared = t * t   # vectorized. element wise squaring
    cost = 1/(2*len(X)) * sum(t_squared)
    return cost
```

```python
def plot(g):
    g2 = g*g
    steps = np.linspace(-4, 6, g)
    w0, w1 = np.meshgrid(steps, steps)
    w0_ = w0.ravel()
    w1_ = w1.ravel()
    w = np.column_stack([w0_, w1_])
    
    def calculateCostArray(X, y, i):
        return computeCost(X, y, w[i])
    
    def computeCost(X, y, w):
        t = X @ w - y
        t_squared = t * t   # vectorized. element wise squaring
        cost = 1/(2*len(X)) * sum(t_squared)
        return cost
    
    costs = np.zeros(g2)
    for i in range(g2):
        costs[i] = calculateCostArray(X_scaled_with_ones_col, y_scaled, i)
    
    costs = costs.reshape(g, g, -1)
    costs = costs.squeeze()
    
    fig3, ax3 = plt.subplots()

    ax3.imshow(costs, origin='lower', extent=(w0.min(), w0.max(), w1.min(), w1.max()))
    ax3.set_title("Cost Function of all the Samples in Figure 1")
    ax3.set_xlabel('$w_0$')
    ax3.set_ylabel('$w_1$')
#     ax3.colorbar()
#     ax3.xlabel('$w_0$')
#     ax3.ylabel('$w_1$')
#     ax3.title('Plot of $cost(\\mathbf{w})$')
#     ax3.gca().set_aspect('equal')  # Force the x and y axes to scale
    

    ax3.contour(w0, w1, costs, colors='white', linestyles=':')


plot(100)
```

this looks good. Upon visual inspection the cost function, I can see that (w0=1.84, w1=2.29) does seem to lie in the darkest part of the cost function. The cost function will be a bowl like shape. Since there's only one feature, I'm aware that there will be only one global minimum. I know this because when I did my hand calculations with just three samples, I've noticed that, the cost function ends up as a 2nd degree polynomial, and that always has a shape of a bowl. Adding more samples to our cost function will still only produce a second degree polynomial, hence one global minimum.

The darkest region within this cost function represents the lowest point of our cost function. We want to let the algorithm know, that has to pay a cost, for the the pair of weights (w0, w1) it picks. Every pair of weights has an associated cost. An algorithm will pick the lowest point on the cost function as this corresponds to the lowest cost.

Now we need to apply gradient descent to our cost function. A gradient descent algorithm with a good learning rate, and sufficient iterations, we need to get w0=1.84 and w1=2.29

### gradient descent

```python
# recall, from the normal equation, we're know 
# we should get w0=1.84 and w1=2.29

normal_equation(X_scaled_with_ones_col, y_scaled)
```

```python
def gradient_descent(X, y, learn_rate=0.00004, num_steps= 10000):
    w = np.array([0, 0]) # so instead of guessing x, we're guessing w
    for i in range(num_steps):
        grad = (X.T @ X) @ w - X.T @ y  # so.. this is a very useful equation
        w = w - learn_rate*grad
    return w

gradient_descent(X_scaled_with_ones_col, y_scaled)
```

```python
# 0.0004 learning rate is quite low and 10000 iterations is quite high
```

```python
w_from_gradient_descent = gradient_descent(X_scaled_with_ones_col, y_scaled)
w_from_gradient_descent
```

we indeed get w0 is 1.84 and w1 is 2.29. the same answer as our normal equation

```python
plot(100)
plt.plot(*w_from_gradient_descent, 'xr')
```
